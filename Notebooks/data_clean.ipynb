# Reviews for every movie in IMDb

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
def get_reviews(id):
    
    reviews_imdb = []
    
    testurl = "https://www.imdb.com/title/tt{}/reviews?ref_=tt_urv".format(id)
    patience_time1 = 0.5
    XPATH_loadmore = "//*[@id='load-more-trigger']"
    XPATH_grade = "//*[@class='review-container']/div[1]"
    list_grades = []

    driver = webdriver.Firefox()
    driver.get(testurl)
    soup = BeautifulSoup(driver.page_source, features="html.parser")
    content = soup.find_all('div', class_=['text','show-more__control'])
    list_content = [tag.get_text() for tag in content]
    reviews_imdb.append(list_content)

    driver.quit()
    
    return reviews_imdb
```
