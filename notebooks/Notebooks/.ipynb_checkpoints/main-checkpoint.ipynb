# 1. Reviews cleaning

```python
import string
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from spacy.lang.en import English

nlp = spacy.load('en')
parser = English()
```

```python
import re

def spacy_tokenizer(sentence):

    tokens = parser(sentence)
    
    filtered_tokens = []
    for word in tokens:
        lemma = word.lemma_.lower().strip()
        
        if lemma not in STOP_WORDS and re.search('^[a-zA-Z]+$', lemma):
            filtered_tokens.append(lemma)

    return filtered_tokens
```

```python
spacy_tokenizer(streaming_platforms["First Review"][1])[:15]
```

# 2. TD-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(min_df=0.25, tokenizer=spacy_tokenizer, ngram_range=(1,2))

tfidf_matrix = tfidf_vectorizer.fit_transform(streaming_platforms["First Review"])

tfidf_matrix.shape
```

```python
terms = tfidf_vectorizer.get_feature_names()

terms[:15]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

dist = 1 - cosine_similarity(tfidf_matrix)

dist.shape
```

# 3. Clustering

# UMAP

```python
import warnings
from umap import UMAP
```

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore')

    umap = UMAP(random_state=42)
    embedding = umap.fit_transform(dist)

embedding[:5]
```

```python
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats

%matplotlib inline
set_matplotlib_formats('svg')

plt.scatter(embedding[:,0], embedding[:,1]);
```

# HDBSCAN

```python
import numpy as np
from hdbscan import HDBSCAN
    
hdbscan = HDBSCAN(min_cluster_size=8)

clustering = hdbscan.fit_predict(embedding)

np.unique(clustering)
```

```python
plt.scatter(embedding[:,0], embedding[:,1], c=clustering);
```
